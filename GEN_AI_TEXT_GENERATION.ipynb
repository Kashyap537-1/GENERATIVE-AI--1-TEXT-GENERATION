#Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?
-   Both Autoencoders (AEs) and Variational Autoencoders (VAEs) utilize an encoder-decoder architecture designed to compress input data into a lower-dimensional latent space before reconstructing it. However, their structural differences drastically impact their utility for text generation:Standard Autoencoders (AEs): An AE maps an input sequence to a precise, fixed point in the latent space. Because there is no constraint on where these points land, the resulting latent space is highly fragmented, containing large "gaps" or unmapped regions. If you sample a random vector from one of these gaps and feed it to the decoder, the model will output gibberish. Consequently, standard AEs are well-suited for tasks like feature extraction or text dimensionality reduction, but they cannot reliably generate new text.Variational Autoencoders (VAEs): Instead of mapping an input to a fixed coordinate, a VAE maps the input to a continuous probability distribution (defined by a mean vector $\mu$ and a variance vector $\sigma$). To ensure the latent space remains continuous and lacks structural gaps, a regularizer called Kullback-Leibler (KL) Divergence is integrated into the loss function. This forces the distributions to cluster cleanly around a standard normal distribution, allowing the decoder to successfully generate coherent, novel text sequences whenever a random point is sampled from the continuous latent space.
#Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical? How the Attention Mechanism Works:Traditional sequence-to-sequence (encoder-decoder) architectures compress an entire source sentence into a single, fixed-size vector.
This "bottleneck" makes it difficult for the model to retain long-range dependencies, often causing it to forget details from the beginning of long sentences.The Attention Mechanism eliminates this bottleneck by allowing the decoder to look back at all of the encoder's intermediate hidden states rather than relying on a single summary vector.Score Calculation: For every word the decoder outputs, it computes an alignment score comparing its current hidden state against all of the encoder's hidden states.Softmax Normalization: These scores pass through a softmax function to produce an array of attention weights that sum up to 1.0.Context Vector Generation: The model calculates a weighted sum of the encoder's hidden states using these attention weights, producing a custom context vector.Prediction: The decoder combines this context vector with its current state to predict the next translated word accurately.Why Attention Mechanisms Are Critical:Mitigating the Information Bottleneck: They ensure long sentences retain key structural details across languages without experiencing performance degradation.Handling Word Alignment Dynamically: Languages possess vastly different syntax layouts (e.g., adjectives preceding nouns in English vs. following nouns in Spanish). Attention allows the model to dynamically shift its focus to the appropriate source words regardless of where they reside in the original sentence.

Question 6: Use the following small text dataset to train a simple Variational
Autoencoder (VAE) for text reconstruction: ["The sky is blue", "The sun is bright", "The grass is
green", "The night is dark", "The stars are shining"] 1. Preprocess the data (tokenize and pad
the sequences). 2. Build a basic VAE model for text reconstruction. 3. Train the model and show
how it reconstructs or generates similar sentences. Include your code, explanation, and sample
outputs. (Include your Python code and output in the code box below.)  

```python
#Answer
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, losses, models
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# 1. Dataset & Preprocessing
data = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

tokenizer = Tokenizer(filters='')
tokenizer.fit_on_texts(data)
vocab_size = len(tokenizer.word_index) + 1
sequences = tokenizer.texts_to_sequences(data)

max_len = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_len, padding='post')

# 2. VAE Architecture Components
latent_dim = 2
embedding_dim = 16
hidden_dim = 32

# Sampling Layer using Reparameterization Trick
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.random.normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# Encoder Architecture
encoder_inputs = layers.Input(shape=(max_len,))
x = layers.Embedding(vocab_size, embedding_dim)(encoder_inputs)
x = layers.LSTM(hidden_dim)(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = models.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

# Decoder Architecture
decoder_inputs = layers.Input(shape=(latent_dim,))
x = layers.Dense(max_len * hidden_dim)(decoder_inputs)
x = layers.Reshape((max_len, hidden_dim))(x)
x = layers.LSTM(hidden_dim, return_sequences=True)(x)
decoder_outputs = layers.TimeDistributed(layers.Dense(vocab_size, activation="softmax"))(x)
decoder = models.Model(decoder_inputs, decoder_outputs, name="decoder")

# Overall VAE Model Wrapper
class VAE(models.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # Reconstruction Loss (Categorical Cross-Entropy)
            recon_loss = tf.reduce_mean(
                losses.sparse_categorical_crossentropy(data, reconstruction)
            ) * max_len

            # KL Divergence Loss
            kl_loss = -0.5 * tf.reduce_mean(
                1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var)
            )
            total_loss = recon_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        return {"loss": total_loss, "recon_loss": recon_loss, "kl_loss": kl_loss}

# 3. Training the Model
vae = VAE(encoder, decoder)
vae.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01))
vae.fit(padded_sequences, epochs=200, batch_size=5, verbose=0)

# 4. Reconstruction Demonstration
print("--- Text Reconstruction Results ---")
z_mean, _, _ = vae.encoder.predict(padded_sequences, verbose=0)
reconstructions = vae.decoder.predict(z_mean, verbose=0)

reverse_word_index = {v: k for k, v in tokenizer.word_index.items()}

for idx, orig_seq in enumerate(padded_sequences):
    orig_text = " ".join([reverse_word_index[w] for w in orig_seq if w != 0])
    recon_text_indices = np.argmax(reconstructions[idx], axis=-1)
    recon_text = " ".join([reverse_word_index[w] for w in recon_text_indices if w != 0])
    print(f"Original:      '{orig_text}'")
    print(f"Reconstructed: '{recon_text}'\n")
```

Question 7: Use a pre-trained GPT model (like GPT-2 or GPT-3) to translate a
short English paragraph into French and German. Provide the original and translated text.
(Include your Python code and output )


```python
#ANSWER

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# Utilizing an instruction-tuned generative LLM (Qwen2.5-1.5B-Instruct) for zero-shot text manipulation
model_name = "Qwen/Qwen2.5-1.5B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype="auto", device_map="auto")

# The original English paragraph
english_text = (
    "Artificial intelligence is transforming the modern world. It helps scientists "
    "discover new medicines, automates repetitive work, and changes how humans interact "
    "with technology daily."
)

def translate_text(text, target_language):
    # Establish a structured prompt instructing the model to translate directly
    messages = [
        {"role": "system", "content": f"You are an expert translator. Translate the given English text into {target_language}. Provide ONLY the raw translation without any conversation or introduction."},
        {"role": "user", "content": text}
    ]

    text_prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    model_inputs = tokenizer([text_prompt], return_tensors="pt").to(model.device)

    generated_ids = model.generate(**model_inputs, max_new_tokens=150, do_sample=False)
    generated_ids = [output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)]

    return tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()

# Generate Translations
french_translation = translate_text(english_text, "French")
german_translation = translate_text(english_text, "German")

# Output Results
print("=================== ORIGINAL ENGLISH TEXT ===================")
print(english_text)
print("\n=================== FRENCH TRANSLATION ===================")
print(french_translation)
print("\n=================== GERMAN TRANSLATION ===================")
print(german_translation)
```

#QUESTION 8: : Implement a simple attention-based encoder-decoder model for English-to-Spanish Translation using Tensorflow or PyTorch.

```python
#ANSWER
import tensorflow as tf
import numpy as np

# 1. Toy Parallel Dataset
input_texts = ['hello .', 'how are you ?', 'i love technology .', 'good morning .', 'welcome home .']
target_texts = ['<start> hola . <end>', '<start> como estas ? <end>', '<start> amo la tecnologia . <end>', '<start> buenos dias . <end>', '<start> bienvenido a casa . <end>']

def tokenize(texts):
    tokenizer = tf.keras.preprocessing.text.Tokenizer(filters='')
    tokenizer.fit_on_texts(texts)
    seqs = tokenizer.texts_to_sequences(texts)
    padded_seqs = tf.keras.preprocessing.sequence.pad_sequences(seqs, padding='post')
    return padded_seqs, tokenizer

input_seq, inp_tokenizer = tokenize(input_texts)
targ_seq, targ_tokenizer = tokenize(target_texts)

# Configuration Parameters
vocab_inp_size = len(inp_tokenizer.word_index) + 1
vocab_tar_size = len(targ_tokenizer.word_index) + 1
embedding_dim = 32
units = 64

# 2. Bahdanau Attention Layer
class BahdanauAttention(tf.keras.layers.Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = tf.keras.layers.Dense(units)
        self.W2 = tf.keras.layers.Dense(units)
        self.V = tf.keras.layers.Dense(1)

    def call(self, query, values):
        query_with_time_axis = tf.expand_dims(query, 1)
        score = self.V(tf.nn.tanh(self.W1(query_with_time_axis) + self.W2(values)))
        attention_weights = tf.nn.softmax(score, axis=1)
        context_vector = attention_weights * values
        context_vector = tf.reduce_sum(context_vector, axis=1)
        return context_vector, attention_weights

# 3. Encoder Model
class Encoder(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, enc_units):
        super(Encoder, self).__init__()
        self.enc_units = enc_units
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(self.enc_units, return_sequences=True, return_state=True)

    def call(self, x, hidden):
        x = self.embedding(x)
        output, state = self.gru(x, initial_state=hidden)
        return output, state

# 4. Decoder Model with Attention Integration
class Decoder(tf.keras.Model):
    def __init__(self, vocab_size, embedding_dim, dec_units):
        super(Decoder, self).__init__()
        self.dec_units = dec_units
        self.embedding = tf.keras.layers.Embedding(vocab_size, embedding_dim)
        self.gru = tf.keras.layers.GRU(self.dec_units, return_sequences=True, return_state=True)
        self.fc = tf.keras.layers.Dense(vocab_size)
        self.attention = BahdanauAttention(self.dec_units)

    def call(self, x, hidden, enc_output):
        context_vector, attention_weights = self.attention(hidden, enc_output)
        x = self.embedding(x)
        x = tf.concat([tf.expand_dims(context_vector, 1), x], axis=-1)
        output, state = self.gru(x)
        output = tf.reshape(output, (-1, output.shape[2]))
        x = self.fc(output)
        return x, state, attention_weights

# Initialize Networks
encoder = Encoder(vocab_inp_size, embedding_dim, units)
decoder = Decoder(vocab_tar_size, embedding_dim, units)
optimizer = tf.keras.optimizers.Adam()
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction='none')

def loss_function(real, pred):
    mask = tf.math.logical_not(tf.math.equal(real, 0))
    loss_ = loss_object(real, pred)
    mask = tf.cast(mask, dtype=loss_.dtype)
    return tf.reduce_mean(loss_ * mask)

@tf.function
def train_step(inp, targ, enc_hidden):
    loss = 0
    with tf.GradientTape() as tape:
        enc_output, enc_hidden = encoder(inp, enc_hidden)
        dec_hidden = enc_hidden
        dec_input = tf.expand_dims([targ_tokenizer.word_index['<start>']] * inp.shape[0], 1)

        for t in range(1, targ.shape[1]):
            predictions, dec_hidden, _ = decoder(dec_input, dec_hidden, enc_output)
            loss += loss_function(targ[:, t], predictions)
            dec_input = tf.expand_dims(targ[:, t], 1)

    batch_loss = (loss / int(targ.shape[1]))
    variables = encoder.trainable_variables + decoder.trainable_variables
    gradients = tape.gradient(loss, variables)
    optimizer.apply_gradients(zip(gradients, variables))
    return batch_loss

# Training Loop
EPOCHS = 250
for epoch in range(EPOCHS):
    enc_hidden = tf.zeros((input_seq.shape[0], units))
    total_loss = train_step(input_seq, targ_seq, enc_hidden)

print(f"Training Complete! Final Cross-Entropy Loss: {total_loss.numpy():.4f}")
```

Question 10: Imagine you are building a creative writing assistant for a publishing company.
The assistant should generate story plots and character descriptions using Generative AI.
Describe how you would design the system, including model selection, training data, bias
mitigation, and evaluation methods. Explain the real-world challenges you might face.  
ANSWER:

```python
import os
import nltk
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM

nltk.download('punkt', quiet=False)

class CreativeWritingAssistant:

    def __init__(self, model_name="gpt2"):
        print(f"[1/4] Model Selection: Initializing core generative engine ({model_name})...")

        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.generator = pipeline('text-generation', model=model_name, tokenizer=self.tokenizer)

        self.bias_blacklist = ["stereotypical", "cliche", "subservient", "fanatic"]

    def mitigate_bias(self, text):

        lowered_text = text.lower()
        for word in self.bias_blacklist:
            if word in lowered_text:

                text = text.replace(word, "[attribute adjusted for diversity alignment]")
        return text

    def generate_creative_artifact(self, prompt_type, seed_idea):

        if prompt_type == "plot":
            prompt = f"Write a compelling story plot based on this concept: {seed_idea}\nPlot Summary:"
        elif prompt_type == "character":
            prompt = f"Create a detailed character profile based on this concept: {seed_idea}\nCharacter Description:"
        else:
            raise ValueError("Invalid prompt_type. Choose 'plot' or 'character'.")


        outputs = self.generator(
            prompt,
            max_new_tokens=150,
            num_return_sequences=1,
            temperature=0.7,
            do_sample=True,
            pad_token_id=self.tokenizer.eos_token_id
        )

        raw_generation = outputs[0]['generated_text']


        clean_generation = self.mitigate_bias(raw_generation)
        return clean_generation

    def evaluate_output(self, generated_text):
        """
        Evaluation Framework
        Simulates production telemetry metrics like basic lexical variety (TTR).
        """
        tokens = nltk.word_tokenize(generated_text.lower())

        if not tokens:
            return 0.0

        ttr = len(set(tokens)) / len(tokens)
        return ttr

if __name__ == "__main__":
    print("=== STARTING CREATIVE WRITING ASSISTANT INITIALIZATION ===\n")

    # 1. Initialize System
    assistant = CreativeWritingAssistant()

    # 2. Simulate Training Data Input Concept / Prompt Generation [cite: 59, 60]
    plot_seed = "A detective discovers that time flows backward inside a locked vault."
    character_seed = "A brilliant engineer who can speak to legacy computer servers."

    print("\n[2/4] Generating Story Artifacts...")
    generated_plot = assistant.generate_creative_artifact("plot", plot_seed)
    generated_character = assistant.generate_creative_artifact("character", character_seed)

    print(generated_plot)

    print(generated_character)

    print("\n[3/4] Running Automated Evaluation Suite...")

    plot_score = assistant.evaluate_output(generated_plot)
    char_score = assistant.evaluate_output(generated_character)

    print(f" -> Generated Plot Lexical Diversity Score (TTR): {plot_score:.4f}")
    print(f" -> Generated Character Lexical Diversity Score (TTR): {char_score:.4f}")

    print("\n[4/4] System Architectural Evaluation Complete.")
```
